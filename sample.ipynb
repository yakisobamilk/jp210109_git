## BokehのグラフはGitHub上ではどう表示されるんだろう？

```python
from bokeh.plotting import figure, output_file, show, reset_output, output_notebook

output_notebook()
p = figure(width=500, height=300, 
#           tooltips=TOOLTIPS,
#           x_range=(-1.05, 1.2), y_range=(-1.05, 1.2),
           active_scroll="wheel_zoom")
p.line(x=x, y=np.sin(x))
p.circle(x=x, y=np.sin(x))
show(p)
```
